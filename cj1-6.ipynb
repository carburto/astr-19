```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    y = np.sin(x)
    return y

#returns sin(x) as function f(x)
```

```python
def g(x):
    y = np.cos(x)
    return y

#returns cos(x) as function g(x)
```

```python
for i in np.arange(0, 2*np.pi, 2*np.pi/1000):
    print('{0:.4f} {1:.4f} {2:.4f}'.format(i, f(i), g(i)))

#tabulates sin and cos of x with 1000 entries
```

```python
for i in np.arange(0, 2*np.pi/100, 2*np.pi/1000):
    print('{0:.4f} {1:.4f} {2:.4f}'.format(i, f(i), g(i)))

#prints first ten of 1000 values of x, sin(x), and cos(x) from 0 to 2pi
#There's probably a simpler way to do this but this is what I've come up with after missing several days of class cause I was sick
```
